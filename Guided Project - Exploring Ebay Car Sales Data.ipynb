```python
# converting the 'price' and 'odometer' columns to numeric values
autos["price"] = (autos["price"]
                 .str.strip()
                 .str.replace("$", "")
                 .str.replace(",", "")
                 .astype(int)
                 )

autos["odometer"] = (autos["odometer"]
                    .str.strip()
                    .str.replace("km", "")
                    .str.replace(",", "")
                    .astype(int)
                    )

autos.rename({"odometer": "odometer_km"}, axis=1, inplace=True)
```

```python
autos["brand"].unique()
```

```python
autos["brand"].value_counts().sort_index()
```

```python
brands = autos["brand"].value_counts().sort_index().index
brands
```

As we can see all values in the brand column are brands with the exception of sonstige_autos, which translates to 'other cars'. I've decided to look only at the brands and calculate the mean price for each brand. I've sorted the values and created an object to hold the brand names to aggregate by, but first I have to remove the sonstige_autos label.

```python
brands = brands.drop("sonstige_autos")
```

```python
# lets aggregate

mean_price_brand = {}

for b in brands:
    b_rows = autos[autos["brand"] == b]
    b_mean = int(b_rows["price"].mean())
    mean_price_brand[b] = b_mean

mean_price_brand
```

A quick analysis of these numbers tells us that asian cars are cheapest, followed by European cars. German cars seem quite decently priced, with Porsche as the exception of course. Landrover also sticks out with a mean price over double that of nearly all other brands. It must be noted that I did not adjust anything to take into account the difference in the number of listings per brand.

## Let's make it a little more interesting before we move on

I'm going to first select the top 5 brands, then calculate the mean mileage and mean price and store the values in seperate dictionairies. From there I'll convert both to series objects and then to a combined dataframe object so I can analyse them next to each other.

```python
top_5_brands  = (autos["brand"]
                 .value_counts()
                 .head(5)
                 .index
                )

top_5_brands
```

```python
mean_price_top_5 = {}

for b in top_5_brands:
    b_rows = autos[autos["brand"] == b]
    b_mean = int(b_rows["price"].mean())
    mean_price_top_5[b] = b_mean

mean_mileage_top_5 = {}

for b in top_5_brands:
    b_rows = autos[autos["brand"] == b]
    b_mean = int(b_rows["odometer_km"].mean())
    mean_mileage_top_5[b] = b_mean

mean_price_top_5
```

```python
mean_mileage_top_5
```

```python
# coverting to series objects
mpt_5_series = pd.Series(mean_price_top_5)
mpt_5_series
```

```python
mmt_5_series = pd.Series(mean_mileage_top_5)
mmt_5_series
```

```python
top_5_mean_df = pd.DataFrame(mpt_5_series, columns=["mean_price"])
top_5_mean_df
```

```python
top_5_mean_df["mean_mileage"] = mmt_5_series
```

```python
print(top_5_mean_df)
```

There seems to be no correlation between the mean mileage and mean price of a specific brand of car on this site. The mean mileage is around 130,000 km (yes, expressing mileage in kilometers is strangely ironic) for all brands, eventhough the mean price strongly differs.
